<h1><Center>Importing All Required Packages

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import difflib
```

<h1><center>Reading The DataSet

```python

data = pd.read_csv('movies.csv',index_col='index')
data
```

<h1><center>Info About DataSet

```python
data.info()
```

<h1><center>Selecting The Required Features 

```python
features = data[['genres','keywords','overview','tagline','title','cast','director']]
features = features.fillna('')
features
```

<h1><center>Combining the Features 

```python
combined_features = features.genres+features.keywords
combined_features
```

<h1><center>Finding the TfidfVectorizer Values for each movie 

```python
vector = TfidfVectorizer()
feature_vectors = vector.fit_transform(combined_features)
print(feature_vectors)
```

<h1><center> Finding the cosine similarities score 

```python
similerity = cosine_similarity(feature_vectors)
similerity
```

<h1><center>Getting the Input Keywords about the Movie form the User 

```python
input_text = input('Search here ')
input_text
```

<h1><center>Finding the Best Match to the Input Keywords

```python
title_match = difflib.get_close_matches(input_text,features.title)
genres_match = difflib.get_close_matches(input_text,features.genres)
keywords_match = difflib.get_close_matches(input_text,features.keywords)
overview_match = difflib.get_close_matches(input_text,features.overview)
tagline_match = difflib.get_close_matches(input_text,features.tagline)
cast_match = difflib.get_close_matches(input_text,features.cast)
director_match = difflib.get_close_matches(input_text,features.director)
```
