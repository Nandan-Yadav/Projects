<h1><Center>Importing All Required Packages

```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import difflib
```

<h1><center>Reading The DataSet

```python

data = pd.read_csv('/Users/nandanyadav/Desktop/programming/divyaproject/movies.csv',index_col='index')
data
```

<h1><center>Info About DataSet

```python
data.info()
```

<h1><center>Selecting The Required Features 

```python
features = data[['genres','keywords','overview','tagline','title','cast','director']]
features = features.fillna('')
features
```

<h1><center>Combining the Features 

```python
combined_features = features.genres+features.keywords+features.overview+features.tagline+features.title+features.cast+features.director
combined_features
```

<h1><center>Finding the TfidfVectorizer Values for each movie 

```python
vector = TfidfVectorizer()
feature_vectors = vector.fit_transform(combined_features)
print(feature_vectors)
```

<h1><center> Finding the cosine similarities score 

```python
similerity = cosine_similarity(feature_vectors)
similerity
```

<h1><center>Getting the Input Keywords about the Movie form the User 

```python
input_text = input('Search here ')
input_text
```

<h1><center>Finding the Closest Matchs to the Input Title

```python
title_match = difflib.get_close_matches(input_text,features.title)
title_match
```

<h1><center>Findinf The Best Closest Match to the Input Title

```python
title_match[0]
```

<h1><center>Finding The Index Of The Movie Title

```python
index_of_the_movie = data[data.title==title_match[0]].index.values[0]
index_of_the_movie
```

<h1><center>Finding The List Of Similarity Score For The Movies

```python
similarity_score = list(enumerate(similerity[index_of_the_movie]))
similarity_score
```

<H1><center>Sorting The Movies Based On The Similarity Score Of That Movie

```python
sort_similarty_movies  = sorted(similarity_score,key=lambda x:x[1],reverse=True)
sort_similarty_movies
```

<H1><center>Similar Movies Based On The Score

```python
# Printig Top 30 Movies
top_30_movie_index = [i[0] for i in sort_similarty_movies[0:30]]
for i in top_30_movie_index:
    print(data.iloc[i,17])
```
